```python
from prophet import Prophet
```

```python
df_main['ds']=pd.to_datetime(df_main['Datetime'])
df_main['y'] =df_main['St(kVA)_update']
```

```python
df_main.head()
```

```python
df_main.drop(columns=['Datetime','St(kVA)_update'],axis=1,inplace=True)
```

```python
df_main.head()
```

```python
df_main.tail()
```

```python
model= Prophet()
```

```python
model.fit(df_main)
```

```python
future = model.make_future_dataframe(periods=30,freq='T')
forecast = model.predict(future)
forecast.tail()
```

```python
forecast =forecast[['ds','yhat']]
```

```python
forecast.shape
```

```python
df_main.shape
```

```python
dataset =pd.merge(forecast,df_main,on ='ds',how='left')
```

```python
dataset['Predicted']=dataset['yhat']
dataset['Actual']=dataset['y']
```

```python
dataset=dataset[['ds','Predicted','Actual']]
```
